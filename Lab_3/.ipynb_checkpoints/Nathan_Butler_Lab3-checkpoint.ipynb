```python
#### Q1: Normalize the dataset using Min-Max normalization
```

```python
column = [
    "Year",
"Days with AQI",
"Good Days",
"Moderate Days",
"Unhealthy for Sensitive Groups Days",
"Unhealthy Days",
"Very Unhealthy Days",
"Hazardous Days",
"Max AQI",
"90th Percentile AQI",
"Median AQI",
"Days CO",
"Days NO2",
"Days Ozone",
"Days SO2",
"Days PM2.5",
"Days PM10"
]


def min_max_scale(data, target_col):
    data = data.copy()
    data_mean = np.mean(data[target_col])
    MAX = np.max(data[target_col])
    MIN = np.min(data[target_col])
    mean_norm = ( (data[target_col] - MIN) / (MAX-MIN) )
    data[target_col] = mean_norm
    return data
min_max = min_max_scale(data, column)
min_max
```

```python
#### Q2: Standardize the dataset using Z-score standardization
```

```python
column = [
    "Year",
"Days with AQI",
"Good Days",
"Moderate Days",
"Unhealthy for Sensitive Groups Days",
"Unhealthy Days",
"Very Unhealthy Days",
"Hazardous Days",
"Max AQI",
"90th Percentile AQI",
"Median AQI",
"Days CO",
"Days NO2",
"Days Ozone",
"Days SO2",
"Days PM2.5",
"Days PM10"
]
def standardize(data, target_col):
    data = data.copy()
    for col in data[target_col]:
        z_score = (data[col] - data[col].mean()) / data[col].std(ddof=0)
        data[col] = z_score
    return data
z_scores = standardize(data, column)
z_scores
```

### 2. Encoding Categorical Variables

#### Q4: Convert the categorical variable State into a numerical form using one-hot encoding. How many new columns are created, and what are their names?

```python
unique_list=list()
for unique in data['State'].unique():
    unique_list.append(unique)
```

```python
def one_hot_encoding(data, column_name, column_values):
    data = data.copy()
    # Create binary columns only for string-type categories
    for dwelling in column_values:
        data[f"{dwelling}"] = (data[column_name] == dwelling).astype(int)
    return data
one_hot = one_hot_encoding(data, 'State', unique_list)
print("There are now 73 columns! Listed below are their names:")
for x in one_hot.columns:
    print(x)
```

#### Q5: Use label encoding for the County column. What are the numerical labels assigned to each county?

```python
def label_encoding(data):
    data = data.copy()
    # We create a map to hold the idx, and unique case values.
    loc_map = {} 
    for idx, case in enumerate(data['County'].unique()):
        loc_map[case] = idx
    data['County'] = data['County'].map(loc_map)
    return data
state_label = label_encoding(data)
print("For each county, the numerical labels range from 0 to 818.")
state_label['County']
```

#### Q6: Discuss the impact of using one-hot encoding vs label encoding on the model performance when the categorical variables are used in regression analysis.

>##### If we are talking about model performance, label encoding would be the best fit in this dataset because we can make predictions on the counts of each County. For the State column, one-hot encoding works well here because with regression analysis there we would try and target those features.

### 3. Discretization and Binning

#### Q7: Bin the Max AQI column into 4 categories: "Low", "Moderate", "High", and "Very High" using equal-width binning. How many rows fall into each category?

```python
import math
minn=round(min(z_scores['Max AQI']))
very_high=max(z_scores['Max AQI'])
moderate=np.mean(z_scores['Max AQI'])
high=(very_high - moderate)
minn
```

```python
import math
def id_outlier(z_scores):
    
    minn=round(min(z_scores['Max AQI']))
    very_high=round(max(z_scores['Max AQI']))
    moderate=round(math.exp(np.mean(z_scores['Max AQI'])))
    high=(very_high - moderate)
    
    for col, scores in z_scores['Max AQI'].items():
        Q1 = scores[int(scores) < moderate] # Low
        Q2 = scores[(int(scores) >= abs(moderate)) & (int(scores) < high)] # Moderate
        Q3 = scores[(int(scores) >= abs(moderate)) & (int(scores) < very_high)] # High
        Q4 = scores[(int(scores) > high) & (int(scores) <= very_high)] # Very High
        z_scores["Q1"] = Q1
        z_scores["Q2"] = Q2
        z_scores["Q3"] = Q3
        z_scores["Q4"] = Q4
    return z_scores
binning = id_outlier(z_scores)
binning
```

```python
#### Q9: Apply a log transformation to the Max AQI column to handle its skewness. How does this affect the skewness?
```

```python
def log_trans(data, target_col):
    data = data.copy()
    
    data[target_col] = np.log(data[target_col])
    
    return data

logg = log_trans(data, 'Max AQI')
logg
# Use matplotlib func
```

```python
#### Q10: After applying the log transformation, what changes do you observe in the distribution of the Max AQI data?
```

```python
def LOG_AQI(data):
    plt.figure(figsize=(8,6))
    plt.bar(data['Max AQI'], height=logg['Max AQI'].values)
    plt.show()
    
LOG_AQI(logg)
print("After applying the log transformation, the distribution is more uniform.")
```
